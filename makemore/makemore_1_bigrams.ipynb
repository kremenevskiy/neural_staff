```python
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
```

```python
words = open('names.txt', 'r').read().splitlines()
print('len: ', len(words))
print('min len: ', min(len(w) for w in words))
print('max len: ', max(len(w) for w in words))
print(words[:5])
```

# Start

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
sorted(b.items(), key=lambda kv: kv[1], reverse=True)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s: i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i: s for s, i in stoi.items()}
```

# BiGram algo

```python
N = torch.zeros((27, 27), dtype=torch.int32)
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
print(N.shape)
```

## Bigram matrix

```python
plt.figure(figsize=(18, 18))
plt.imshow(N, cmap='Blues')
for i in range(27):
    for j in range(27):
        chstr = itos[i] + '->' + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
        plt.text(j, i, N[i, j].item(), ha="center", va="top", color="gray")
plt.axis('off');
```

```python
P = N.float()
P = (N+1).float() # smoothing model like regularization model
# P = P / P.sum(dim=1, keepdim=True)
P /= P.sum(dim=1, keepdim=True) # faster 
# 27 27
# 27 1   
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(10):
    out = []
    ix = 0
    while True:
        p = P[ix]
#         p = torch.ones(27) / 27.0 # random suggestions
        
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

# LogLikelyHood

```python
# log(a * b * c) = log(a) + log(b) + log(c)
```

```python
# GOAL: maximize likelihood of the data w.r.t. model parameters (statistical modeling)
# Maximize probabilities assigned by the model
# We want to change parameters of the model to maximize this probabilites


# We see what probability was assigned by the model that ch2 go after ch1
# and we want that very high so we tune parameters

# Maximizing likelihood like maximizing log_likelihood
# log(a * b * c) = log(a) + log(b) + log(c) 
```

```python
log_likelihood = 0.0
n = 0
for w in words[:3]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
        
print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
print(f'{nll/n=}')
```

```python
nlls = torch.zeros(5)
for i in range(5):
    # i-th bigram:
    x = xs[i].item() # input character index
    y = ys[i].item() # label character index
    print('--------')
    print(f'bigram example {i+1}: {itos[x]}{itos[y]} (indexes {x},{y})')
    print('input to the neural net:', x)
    print('output probabilities from the neural net:', probs[i])
    print('label (actual next character):', y)
    p = probs[i, y]
    print('probability assigned by the net to the the correct character:', p.item())
    logp = torch.log(p)
    print('log likelihood:', logp.item())
    nll = -logp
    print('negative log likelihood:', nll.item())
    nlls[i] = nll

print('=========')
print('average negative log likelihood, i.e. loss =', nlls.mean().item())
```

## Train NN of Full dataset

```python
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
        
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
print('number of examples: ', num)

# Initialize the 'network'
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((27, 27), generator=g, requires_grad=True) # W from here as N from begram statistic, by it is log N
# W.exp() is N

losses = []
```

```python
# gradient descent
for k in range(1000):
    # Forward pass
    xenc = F.one_hot(xs, num_classes=27).float() # input to the network: one-hot encoding
    logits = xenc @ W # predict log-counts
    counts = logits.exp() # counts, equivalent to N
    probs = counts / counts.sum(1, keepdims=True) # probabilities for next character
    loss = -probs[torch.arange(num), ys].log().mean() # + 0.01*(W**2).mean() # out loss
    losses.append(loss.item())
    
    # backward pass
    W.grad = None # set grad to zero
    loss.backward()
    
    # update parameters
    h = 50
    W.data += -h * W.grad
    
print('loss: {:.4f}'.format(loss.item()))
print('loss from counting 2.4541')
```

```python
plt.plot(losses)
plt.title('loss by iters')
```

## Sample from 'neural net' model

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(10):
    out = []
    ix = 0
    
    while True:
        xenc = F.one_hot(torch.tensor([ix]), num_classes=27).float()
        logits = xenc @ W # predict log-counts
        counts = logits.exp() # counts, equivalent to N
        p = counts / counts.sum(1, keepdim=True) # probabilities for next character
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
